```
import networkx as nx
import scipy as sp
import numpy as np

data = np.loadtxt('facebook_combined.txt')
G = nx.Graph()
for each in data:
  G.add_edge(int(each[0]),int(each[1]))
```

```
degreeval = G.degree()  
degreeval
```

```
deg = np.array(degreeval)
deg
```

```
degx = deg[:,0]
degy = deg[:,1]
degx
degy
```

```
kmini = np.min(degy)
kmini
```

```
N = len(degy)
N
```

```
import matplotlib.pyplot as plt            
degree =  nx.degree_histogram(G)         
x = range(len(degree)-10)   
print(x)
#print(len(degree))      
degree=degree[10:]          #生成x轴序列，从1到最大度

y = [z / float(sum(degree)) for z in degree] 
#print(y)


a = np.linspace(5, 600, 600)  


plt.plot(x,y,color="red",linewidth=1,linestyle=':') 
plt.show() 
#x1 = -1.3*np.log(degy)


#plt.loglog(x,y,color="orange",linewidth=1,linestyle=':')  

#plt.plot(np.log(deg[:,0]),-alp*np.log(deg[:,1]+c))
#a=x

plt.loglog(x,y,color="orange",linewidth=1,linestyle=':')  
plt.loglog(a,0.15*a**(-0.9),color="blue",linewidth=1)
plt.show() 
```
